```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal, constants
```

```python
Tau=1200
m=1
alpha=(m*np.pi)/Tau
t=np.arange(0,Tau)
sq_pulse=np.zeros(Tau)
sq_pulse[int(Tau/3):int(Tau*2/3)]=1
plt.plot(sq_pulse)
sin_pulse=np.sin(alpha*t)
plt.plot(sin_pulse)
sinsq_pulse=(np.sin(alpha*t))**2
plt.plot(sinsq_pulse)
```

```python
sq_fft=np.fft.fft(sq_pulse)
plt.plot(sq_fft.real,sq_fft.imag)
#plt.plot(np.abs(sq_fft))

sin_fft=np.fft.fft(sin_pulse)
plt.plot(sin_fft.real,sin_fft.imag)

sinsq_fft=np.fft.fft(sinsq_pulse)
plt.plot(sinsq_fft.real,sinsq_fft.imag)
```
